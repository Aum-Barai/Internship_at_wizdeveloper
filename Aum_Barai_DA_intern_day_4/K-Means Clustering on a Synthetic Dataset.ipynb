```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Generate synthetic dataset with more realistic variance and distribution
n_samples = 300
n_features = 2
n_clusters = 3
cluster_std = [3.0, 2.5, 3.5]  # Different standard deviations for each cluster
X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, cluster_std=cluster_std, random_state=42)
```

```python
# Adding more complex noise to make the dataset more realistic
noise = np.random.normal(0, 0.5, X.shape)  
X += noise
```

```python
# Plot the synthetic dataset
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, alpha=0.6, edgecolor='k')
plt.title("Synthetic Dataset with Realistic Variance and Noise")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
# Plot the clustered data
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', alpha=0.6, edgecolor='k')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', edgecolor='k')
plt.title("K-Means Clustering with Enhanced Realistic Data and Noise")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
